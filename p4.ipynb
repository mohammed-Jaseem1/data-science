```python
import matplotlib.pyplot as plt
import numpy as np 
plt.suptitle("graph", fontsize=16)

xpoint = np.array([2,5])
ypoint = np.array([2,8])

plt.subplot(3,3,1)
plt.title("title")
plt.xlabel("blue color")
plt.ylabel("blue color")
plt.plot(xpoint,ypoint,'o')
plt.tight_layout()



x1point = np.array([0, 1, 2, 3])
y1point = np.array([3, 8, 1, 10])

x2point = np.array([0, 1, 2, 3])
y2point = np.array([6, 4, 6, 3])
plt.subplot(3,3,2)

plt.plot(x1point,y1point,linestyle='dotted')
plt.plot(x2point,y2point,marker='*',linewidth='2.5')
plt.tight_layout()


plt.subplot(3,3,2)



x1point = np.array([0, 1, 2, 3])
y1point = np.array([3, 8, 1, 10])
plt.subplot(3,3,3)
plt.plot(x1point,y1point,marker='o',ms=10,mfc='yellow',mec='black')
plt.tight_layout()

xpoint = np.array([2,5])
ypoint = np.array([2,8])

plt.subplot(3,3,4)
plt.plot(xpoint,ypoint,)
plt.grid()
plt.subplot(3,3,5) 

x3 = np.array([5,7,8,7,2])
y3 = np.array([99,86,87,88,111])
plt.scatter(x3, y3)
x4 = np.array([6,7,10,3,8])
y4 = np.array([55,87,88,83,101])
plt.scatter(x4, y4,color='r')
plt.subplot(3,3,5) 
plt.tight_layout()

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.subplot(3,3,6)
plt.bar(x,y,color='r',width=0.1)
plt.tight_layout()

y = np.array([35, 20, 20, 25])
mylabels = ["Explode", "Shadow", "Color", "Legend"]
explode = [0.1, 0, 0, 0]  
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'] 
plt.figure(figsize=(10, 8)) 
plt.subplot(3, 3, 7)
plt.pie(y,
        labels=mylabels,
        explode=explode,
        shadow=True,
        colors=colors,
        autopct='%1.1f%%')  
plt.title("Pie Chart with All Features")
plt.legend(title="Labels", loc="upper right")  

plt.tight_layout()

plt.show()
```
